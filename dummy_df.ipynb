```python
import pandas as pd
import numpy as np
import random
```

#### Values for column "Spend"

```python
rows = 240 # vamos a asignar 20 d√≠as a cada mes, por eso el 240
total_spend = 36000000
n_random = np.random.rand(rows)
spend = n_random/n_random.sum() * total_spend
```

#### Values for column "Video Views"

```python
total_views = 93000
n_random = np.random.rand(rows)
video_views = n_random/n_random.sum() * total_views
```

#### Values for column "Impressions"

```python
total_impressions = 89000
n_random = np.random.rand(rows)
impressions = n_random/n_random.sum() * total_impressions
```

#### Values for column "Clicks"

```python
total_clicks = 1200
n_random = np.random.rand(rows)
clicks = n_random/n_random.sum() * total_clicks
```

#### Values for column "Purchase"

```python
total_purchases = 791
n_random = np.random.rand(rows)
purchases = n_random/n_random.sum() * total_purchases
```

#### Values for column "Date"

```python
start = np.datetime64('2023-01-01')
end = np.datetime64('2023-12-31')
date = np.random.choice(np.arange(start, end + np.timedelta64(1, 'D')), size=rows, replace=False)
```

#### Lists of categorical values for columns "Channel", "Source" and "Campaign"

```python
channel_list = ['Programmatic','Paid Search','Paid Social','Organic']
source_list = ['Amazon Ad Server','StackAdapt','LinkedIn Ads','Facebook','Google Display & Video360',
              'Bing Ads', 'Google Search Ads 360']
campaign_list = ['Business-focused zero tolerance','persistent 24/7 attitude','Integraded dedicated contingency',
                'Profound intangible policy','Centralized modular throughput','Automated uniform software',
                'Cross-platform static hierarchy','Networked value-added time-frame']

channel = np.random.choice(channel_list, size=rows)
source = np.random.choice(source_list, size=rows)
campaign = np.random.choice(campaign_list, size=rows)
```

#### Creating a dataframe

```python
dict_df = {
    'spend': spend,
    'video views': np.round(video_views),
    'impressions': np.round(impressions),
    'clicks': np.round(clicks),
    'purchases': np.round(purchases),
    'channel': channel,
    'source': source,
    'campaign': campaign,
    'date': date
}
df = pd.DataFrame(dict_df)
df.head()
```

```python
df.to_csv('dataframe.csv', index=False)
```
